```python
import numpy as np
```

```python
def gradient_descent(x,y):
  m_current=b_current=0
  iterations=1000
  n=len(x)
  learning_rate=0.001 
# cost function is (1/n)*sum(yi-y_predicted)^2  
#md and bd are dertivatives w.r.t m and b of cost function
  for i in range(iterations):
    y_predicted=m_current*x+b_current 
    
      
    cost=(1/n)*sum([val**2 for val in ( y-y_predicted) ])
    md=-(2/n)*sum(x*(y-y_predicted))
    bd=-(2/n)*sum(y-y_predicted)
    m_current=m_current-learning_rate*md
    b_current=b_current-learning_rate*bd
    print("m{} , b{} , cost {}, iterations{}".format(m_current,b_current,cost,iterations))


x=np.array([1,2,3,4,5])
y=np.array([5,7,9,11,13])
gradient_descent(x,y)


```

```python
a0_values=[]
cost_values=[]
a1_values=[]

def Gradient_Descent(X,Y):
  l_rate=float(input("Enter the learning rate  "))
  iter_rations=input("Enter the iterations you want in the algorithm")
  a_0=0
  a_1=0
  n=len(X)
 
  for i in range(int(iter_rations)):
      Y_pred=a_0+a_1*X
      cost =(1/n)*sum([value**2 for value in (Y-Y_pred)])
      partial_a0=(-2/n)*sum(X*(Y-Y_pred))
      partial_a1=(-2/n)*sum(Y-Y_pred)
      a_0=a_0-l_rate*partial_a0
      a_1=a_1-l_rate*partial_a1
      a0_values.append(a_0)
      cost_values.append(cost)
      a1_values.append(a_1)

   
   
    
      print("a0{} , a1{} , cost{} , iterations{}, ".format(a_0,a_1,cost,iter_rations))

```
