```python
import pandas as pd
from statsmodels.stats.outliers_influence import variance_inflation_factor
def calculate_vif(dataset):
  vif=pd.DataFrame()
  vif["features"]=dataset.columns
  vif["vif values"]=[variance_inflation_factor(dataset.values,i) for i in range(dataset.shape[1])]
  return vif
```

```python
df=pd.read_csv("/content/sample_data/california_housing_test.csv")
calculate_vif(df)
```

```python
# lrdf=linear regression dataframe
import pandas as pd
df=pd.read_csv("/content/sample_data/california_housing_test.csv")
df.head(4)
x=df["population"].values
y=df["households"].values
print(x)
import numpy as np
import matplotlib.pyplot as plt
x_mean=np.mean(x)
y_mean=np.mean(y)
num=0
den=0
for i in range(len(x)):
  num +=(x[i]-x_mean)*(y[i]-y_mean)
  den +=(x[i]-x_mean)**2

m=num/den
print(f"m slope value is{m}")
c=y_mean-m*x_mean
print(f"c intercept value is {c}")
y_pred=c+m*x
print(y_pred)
plt.plot(x,y_pred,c="red")
plt.scatter(x,y,c="blue") 
plt.show()
```

```python
from sklearn import linear_model
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("/content/sample_data/california_housing_test.csv")
x=df["population"].values.reshape(-1,1)
y=df["households"].values
ols=linear_model.LinearRegression()
model=ols.fit(x,y)
response=model.predict(x)
r2=model.score(x,y)
coeff=model.coef_
intercept=model.intercept_
print(f"response is{response}")
print(f"r2 is{r2}")
print(f"coeff is{coeff}")
print(f" intercept is{intercept}")
plt.style.use('ggplot')
plt.plot(x,response)
plt.scatter(x,y,c="yellow")
```

```python

df=pd.read_csv("/content/sample_data/california_housing_train.csv")
df.head(5)
x1=df["total_rooms"].values
x2=df["total_bedrooms"].values
y=df["population"].values

```

```python
import numpy as np
def multiple_regression(x1,x2,y):
  sum_x1_square=sum(x1*x1)-(sum(x1)*sum(x1))/len(x1)
  sum_x2_square=sum(x2*x2)-(sum(x2)*sum(x2))/len(x2)
  sum_x1_x2=    sum(x1*x2)-(sum(x1)*sum(x2))/len(x1)
  sum_x1_y=    sum(x1*y)-(sum(x1)*sum(y))/len(x1)
  sum_x2_y=    sum(x2*y)-(sum(x2)*sum(y))/len(x2)

  x1_mean=np.mean(x1)
  x2_mean=np.mean(x2)
  y_mean=np.mean(y)
  b_1=((sum_x2_square)*(sum_x1_y)-(sum_x1_x2)*(sum_x2_y))/((sum_x2_square)*(sum_x1_square)-(sum_x1_x2)*(sum_x1_x2))
  b_2=((sum_x1_square)*(sum_x2_y)-(sum_x1_x2)*(sum_x1_y))/((sum_x1_square)*(sum_x2_square)-(sum_x1_x2)*(sum_x1_x2))
  b_0=y_mean-(b_1*x1_mean)-(b_2*x2_mean)
  print(f"b1 value is {b_1}")
  print(f"b0 value is {b_0}")
  print(f"b2 value is {b_2}")

multiple_regression(x1,x2,y)
```
